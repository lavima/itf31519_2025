```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeRegressor, plot_tree
```

```python
xs = [2,3,5,6,8]
ys = [2,4,5,4,7]

fig = plt.figure()
plt.ylim([0,8])
plt.xlim([0,10])
plt.scatter(xs,ys)
plt.show()
```

```python
split = (xs[0]+xs[1])/2.0
avg1 = ys[0]
avg2 = sum(ys[1:])/len(ys[1:])

fig = plt.figure()
plt.ylim([0,8])
plt.xlim([0,10])
plt.scatter(xs,ys)
plt.plot([split, split], [0, 10], color='green')
plt.plot([0,split],[avg1, avg1], color='red')
plt.plot([split,10],[avg2, avg2], color='red')
plt.show()
```

```python
ssr2 = sum([(y-avg2)**2 for y in ys[1:]])
ssr = [ssr2]

fig = plt.figure()
plt.ylim([2,10])
plt.xlim([-1,5])
plt.scatter(range(0,len(ssr)),ssr)
plt.savefig('ssr1.png')
plt.show()
```

```python
split = (xs[1]+xs[2])/2.0
avg1 = sum(ys[0:2])/len(ys[0:2])
avg2 = sum(ys[2:])/len(ys[2:])

fig = plt.figure()
plt.ylim([0,8])
plt.xlim([0,10])
plt.scatter(xs,ys)
plt.plot([split, split], [0, 10], color='green')
plt.plot([0,split],[avg1, avg1], color='red')
plt.plot([split,10],[avg2, avg2], color='red')
plt.show()
```

```python
ssr1 = sum([(y-avg1)**2 for y in ys[0:2]])
ssr2 = sum([(y-avg2)**2 for y in ys[2:]])
print(ssr)
ssr.append(ssr1+ssr2)
print(ssr)

fig = plt.figure()
plt.ylim([2,10])
plt.xlim([-1,5])
plt.scatter(range(0,len(ssr)),ssr)
plt.show()
```
